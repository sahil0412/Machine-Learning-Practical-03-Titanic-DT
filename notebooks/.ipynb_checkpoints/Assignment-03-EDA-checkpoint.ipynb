```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
import sklearn
from pandas import Series, DataFrame
from pylab import rcParams
from sklearn import preprocessing
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import train_test_split
from sklearn import metrics
from sklearn.metrics import classification_report
```

```python
url = "https://raw.githubusercontent.com/BigDataGal/Python-for-Data-Science/master/titanic-train.csv"
data = pd.read_csv(url)
```

```python
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

### Null values are present

##  We can modify these column as well

```python
data["Has_Cabin"] = (data["Cabin"].notnull().astype('int'))
```

```python
data = data.drop(['Cabin'], axis=1)
```

```python
data.head()
```

```python
data['Embarked'] = data['Embarked'].fillna('S')
```

```python
data.head()
```

```python
# If you want to fill missing values with the median instead
median_value = data['Age'].median()
data['Age'] = data['Age'].fillna(median_value)
```

```python
data.isnull().sum()
```

## We can add SibSp and Parch column as well since both represent no of persons aboarded more

```python
data["no_of_persons"] = data["SibSp"] + data["Parch"]
```

```python
data.head()
```

## Drop other two columns

```python
data = data.drop(labels=["SibSp", "Parch"], axis=1)
```

#### As per the dataset, It contains some columns which are of no use for any algorithm such as:
1. PassengerId -> A unique identifier for each passenger.
2. Name -> The name of the passenger.
3. Ticket -> The ticket number

```python
data = data.drop(labels=["PassengerId", "Name", "Ticket"], axis=1)
```

```python
data.head()
```

```python
## Check for duplicated records
data.duplicated().sum()
```

## Duplicate value found

```python
data = data.drop_duplicates()
```

## So from here we can conclude that `Survived`, `Has_Cabin` and `Pclass` are not numerical column instead they are converted to numerical feature by default, so need to encode them

```python
print(type(numerical_columns))
```

```python
numerical_columns_list = list(numerical_columns)
numerical_columns_list.remove('Survived')
numerical_columns_list.remove('Has_Cabin')
numerical_columns_list.remove('Pclass')
# Convert the list back to Index
numerical_columns = pd.Index(numerical_columns_list)

# Now numerical_columns will have the modified values
print(numerical_columns)
```

```python
categorical_columns_list = list(categorical_columns)
categorical_columns_list.extend(['Survived','Has_Cabin', 'Pclass'])
# Convert the list back to Index
categorical_columns = pd.Index(categorical_columns_list)

# Now numerical_columns will have the modified values
print(categorical_columns)
```
